## Basic Historgrams of Activities and docking pose scores

```python
plt.figure(figsize=(12,6));
plt.title('Activity');

data = activitiesTrain

plt.hist(data, bins=35);
#essentially most things are 0

print("Mean: ", np.sum(data)/len(data))
print("Std: ", np.std(data))
```

```python
#print(labelsTest[0:10])
allDockingScores = compoundDataTrain[:,0:10]

plt.figure(figsize=(12,6));
plt.title('All Docking Scores');

data = allDockingScores.flatten()

plt.hist(data, bins=50, range=[-10,0]);

print("Mean: ", np.sum(data)/len(data))
print("Std: ", np.std(data))
```

```python
allDockingScores = compoundDataTrain[:,0:10]

plt.figure(figsize=(12,6));
plt.title('Best Docking Scores');

data = np.amax(allDockingScores, axis=1)

plt.hist(data, bins=50, range=[-10,0]);

print("Mean: ", np.sum(data)/len(data))
print("Std: ", np.std(data))
```
